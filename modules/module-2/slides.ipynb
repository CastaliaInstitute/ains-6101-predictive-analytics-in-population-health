# Module 2: Risk stratification and prediction

AINS6101 — Predictive Analytics in Population Health

# Essential Question

How are patients grouped for intervention?

# Scenario

a population health team prioritizing outreach for patients with elevated near-term risk

Stakeholders: population health director, care manager, privacy officer, and community advisory representative

# Core Moves

- Define the decision boundary
- Compare baseline and alternative
- Interpret evidence and assumptions
- Identify failure modes
- Recommend next action

# Lab & Assignment

Build or specify a risk stratification model.

Artifact: risk stratification brief with equity checks, intervention criteria, and monitoring plan focused on risk stratification and prediction: Build or specify a risk stratification model.
